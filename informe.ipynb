```python
import pandas as pd
import os
import numpy as np
#Matplotlib forms basis for visalization in Python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

#Graphics in SVG format are more sharp and legible
%config InlineBackend.figure_format = 'svg'

#Increase the default plot size and set the color scheme
plt.rcParams['figure.figsize'] = 12, 9
plt.rcParams['image.cmap'] = 'viridis'
```

```python
##To know path where we are
print(os.getcwd())
```

```python
%%time
from pandas import ExcelWriter
from pandas import ExcelFile
#Export Excel files
stats = pd.read_excel('ceva418_para_rodrigo.xlsx',sheet_name='ceva418') 
```

```python
###Ventas totales de CEVA en los paises filtrados en USD
```

```python
stats.VALTOT18_cer[(stats.PAIS.isin(["BEL","CR","GT", "HON", "PAN", "SAL"])) & (stats.LAB=='CEVA')].sum()
```

```python
##Cantidad de productos ofrecidos y su ramo
```

```python
stats.DESCRIP[(stats.PAIS.isin(["BEL","CR","GT", "HON", "PAN", "SAL"])) & (stats.LAB=='CEVA')].value_counts()
```

```python
###Filtro de productos porcinos para Myco, APP y circovirus delimitado por laboratorios y paises
```

```python
df = stats.loc[(stats.PAIS.isin(["BEL","CR","GT", "HON", "PAN", "SAL"])) & stats.LAB.isin(["CEVA", "BOEHRINGER", "MSD", "ELANCO"]) &(stats.DESCRIP.str.contains('porcina')) & (stats.DESCRIP.str.startswith('Myco') | (stats.DESCRIP.str.startswith('Circovirus')) | (stats.DESCRIP.str.startswith('Actino'))),['LAB','DESCRIP','PRODUCTO_PRES_1', 'VALTOT18_cer']]
```

```python
#Tabla resultante del filtro
df
```

```python
#Ventas CEVA (USD)
```

```python
df[df['LAB'] == "CEVA"].pivot_table(index=["PRODUCTO_PRES_1"],values=["VALTOT18_cer"], margins = True, aggfunc=np.sum, fill_value=0)
```

```python
df[df['LAB'] == "CEVA"].pivot_table(index=["PRODUCTO_PRES_1"],values=["VALTOT18_cer"], margins = True, aggfunc=np.sum, fill_value=0).plot(kind= 'bar')
plt.ylabel("Ventas CEVA")
```
